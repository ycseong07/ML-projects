# Multi-layer Perceptron

## Basic Perceptron

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        torch.manual_seed(0)
        self.fc1 = nn.Linear(in_features=3, out_features=2)
    def forward(self, x):
        x = self.fc1(x)
        return x
```

```python
#  class inheritance
net = Net()
print(net)
```

```python
# check network parameters
print(list(net.parameters()))
```

```python
# create input
input_tensor = torch.tensor([1, 1, 1], dtype=torch.float)
print(input_tensor)
```

```python
# feed forward
out = net(input_tensor)
print(out)
```

```python
# compute manullay
print(-0.0043 + 0.3097 - 0.4752 - 0.0114)
print(-0.4249 - 0.2224 + 0.1548 + 0.4578)
```

```python
# set loss fn. and optimizer
from torch import optim
def criterion(y_hat, y):
  return F.mse_loss(y_hat, y)


optimizer = optim.SGD(net.parameters(), lr=0.01, momentum=0.5)
print(optimizer)
```

```python
# Training loop
data = [torch.tensor([1, 1, 1], dtype=torch.float), torch.tensor([2, 2, 2], dtype=torch.float), torch.tensor([3, 3, 3], dtype=torch.float)]
labels = [torch.tensor([1, 1], dtype=torch.float), torch.tensor([4, 4], dtype=torch.float), torch.tensor([9, 9], dtype=torch.float)]
for epoch in range(20):
  for _, batch in enumerate(zip(data, labels)):
    optimizer.zero_grad()
    x, y = batch[0], batch[1]
    outputs = net(x)
    loss = criterion(outputs, y)
    loss.backward()
    optimizer.step()
  print("Epoch {} - loss: {:.3f} / y_hat: {} - y: {}".format(epoch, loss.item(), outputs.data, y))
```

## Multi-Layer Perceptron

```python
# One-Layer Perceptron

class OneLayerPerceptron(nn.Module):
    def __init__(self):
        super(OneLayerPerceptron, self).__init__()
        torch.manual_seed(0)
        self.fc1 = nn.Linear(1,1)
    def forward(self, x):
        x = self.fc1(x)
        return x

single_net = OneLayerPerceptron()
print(single_net)
```

```python
# Multi-Layer Perceptron

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        torch.manual_seed(0)
        self.fc1 = nn.Linear(1, 5)
        self.fc2 = nn.Linear(5, 1)
    def forward(self, x):
        x = self.fc1(x)
        x = self.fc2(x)
        return x

multi_net = Net()
print(multi_net)
print(list(multi_net.parameters()))
```

```python
# set loss fn
from torch import optim
def criterion(y_hat, y):
  return F.mse_loss(y_hat, y)
```

```python
# create data
data = [(1,3), (2,6), (3,9), (4,12), (5,15), (6,18)]

# set optimizer
optimizer = optim.SGD(single_net.parameters(), lr=0.01, momentum=0.5)

# train loop
for epoch in range(20):
  for _, batch in enumerate(iter(data)):
    x, y = batch[0], batch[1]
    input_tensor = torch.tensor([x], dtype=torch.float)
    label = torch.tensor([y], dtype=torch.float)

    optimizer.zero_grad()
    outputs = single_net(input_tensor)
    loss = criterion(outputs, label)
    loss.backward()
    optimizer.step()

  print("Epoch {} - loss: {:.3f} / y_hat: {} - y: {}".format(epoch, loss.item(), outputs.data, y))
```

```python
# create data
data = [(1,3), (2,6), (3,9), (4,12), (5,15), (6,18)]

# set optimizer
optimizer = optim.SGD(multi_net.parameters(), lr=0.001, momentum=0.5)

# train loop
for epoch in range(20):
  for _, (x, y) in enumerate(iter(data)):
    input_tensor = torch.tensor([x], dtype=torch.float)
    label = torch.tensor([y], dtype=torch.float)

    optimizer.zero_grad()
    outputs = multi_net(input_tensor)
    loss = criterion(outputs, label)
    loss.backward()
    optimizer.step()

  print("Epoch {} - loss: {:.3f} / y_hat: {} - y: {}".format(epoch, loss.item(), outputs.data, label))
```

## MNIST tutorial

```python
import numpy as np
import torch
import torchvision
import matplotlib.pyplot as plt
from time import time
from torchvision import datasets, transforms
from torch import nn, optim
```

```python
# set transform
transform = transforms.Compose([transforms.Resize((28, 28)),
                                transforms.ToTensor(),
                                transforms.Normalize(0.5, 0.5),
                              ])
```

```python
# build dataloader
trainset = datasets.MNIST('../data/CNN_DNN', download=True, train=True, transform=transform)
valset = datasets.MNIST('../data/CNN_DNN', download=True, train=False, transform=transform)

trainloader = torch.utils.data.DataLoader(trainset, batch_size=128, shuffle=True)
valloader = torch.utils.data.DataLoader(valset, batch_size=128, shuffle=False)
```

```python
# check size
dataiter = iter(trainloader)
images, labels = dataiter.next()

print(images.size())
print(labels.size())
```

```python
figure = plt.figure()
num_of_images = 60
for index in range(1, num_of_images + 1):
    plt.subplot(6, 10, index)
    plt.axis('off')
    plt.imshow(images[index].numpy().squeeze(), cmap='gray_r')
```

```python
# Multi-Layer Perceptron

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        torch.manual_seed(0)
        self.model = nn.Sequential(
            nn.Linear(28*28, 128),
            nn.ReLU(inplace=True),
            nn.Linear(128, 64),
            nn.ReLU(inplace=True),
            nn.Linear(64, 10),
        )

    def forward(self, x):
        logit = self.model(x)
        return logit
```

```python
dnn = Net()
print(dnn)
```

```python
# set cost function
criterion = nn.CrossEntropyLoss()
```

```python
# Train the model
optimizer = optim.SGD(dnn.parameters(), lr=0.003, momentum=0.9)
time0 = time()
epochs = 15
for e in range(epochs):
    running_loss = 0
    for images, labels in trainloader:
        # Flatten MNIST images into a 784 long vector
        images = images.view(images.size(0), -1)
    
        # Training pass
        optimizer.zero_grad()
        
        output = dnn(images)
        loss = criterion(output, labels)
        
        # This is where the model learns by backpropagating
        loss.backward()
        
        # And optimizes its weights here
        optimizer.step()
        
        running_loss += loss.item()
    print("Epoch {} - Training loss: {}".format(e, running_loss/len(trainloader)))
    
print("\nTraining Time (in minutes) =",(time()-time0)/60)
```

```python
# Test phase
dnn.eval()
accuracy = []
for images,labels in valloader:
  images = images.view(images.size(0), -1)
  with torch.no_grad():
      logit = dnn(images)

  # Classification Acc.
  _, predicted = torch.max(logit, 1)
  correct = predicted.eq(labels).sum().item()
  batch_acc = 100.*(correct/labels.size(0))
  accuracy.append(batch_acc)

print("\nModel Accuracy =", np.mean(accuracy))
```

##  CIFAR10 Tutorial (2D Conv)

```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from time import time
```

```python
# create dataloader
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

trainset = torchvision.datasets.CIFAR10(root='../data/CNN_DNN', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=32,
                                          shuffle=True, num_workers=4)

testset = torchvision.datasets.CIFAR10(root='../data/CNN_DNN', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=32,
                                         shuffle=False, num_workers=4)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
# Visualize dataset

import matplotlib.pyplot as plt
import numpy as np

# functions to show an image


def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


# get some random training images
dataiter = iter(trainloader)
images, labels = dataiter.next()

# show images
imshow(torchvision.utils.make_grid(images))
# print labels
print(' '.join('%5s' % classes[labels[j]] for j in range(4)))
```

```python
# define a CNN
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.reshape(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x


net = Net()
```

```python
# Define a Loss function and optimizer on CPU

import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
# Train the network on CPU

for epoch in range(2):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        start_time = time()
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 200 == 199:    # print every 200 mini-batches
            print('[{}, {}] loss: {:.3f} time: {:.3f}:'.format
                  (epoch + 1, i + 1, running_loss / 200, (time()-start_time)*200))
            running_loss = 0.0

print('Finished Training')
```

```python
# save the trained model

PATH = '../data/CNN_DNN/cifar_net.pth'
torch.save(net.state_dict(), PATH)
```

```python
# Test the model

dataiter = iter(testloader)
images, labels = dataiter.next()

# print images
imshow(torchvision.utils.make_grid(images))
print('GroundTruth: ', ' '.join('%5s' % classes[labels[j]] for j in range(4)))
```

```python
# load the trained model
#   
net = Net()
net.load_state_dict(torch.load(PATH))
```

```python
# Testing on CPU

correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

print('Accuracy of the network on the 10000 test images: %d %%' % (
    100 * correct / total))
```

```python
# Performance on each class

class_correct = list(0. for i in range(10))
class_total = list(0. for i in range(10))
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = net(images)
        _, predicted = torch.max(outputs, 1)
        c = predicted.eq(labels).squeeze()
        for i in range(4):
            label = labels[i]
            class_correct[label] += c[i].item()
            class_total[label] += 1


for i in range(10):
    print('Accuracy of %5s : %2d %%' % (
        classes[i], 100 * class_correct[i] / class_total[i]))
```

```python
# Model on GPU

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Assuming that we are on a CUDA machine, this should print a CUDA device:

print(device)
```

```python
# Define a Loss function and optimizer on CPU

import torch.optim as optim

criterion = nn.CrossEntropyLoss()
```

```python
net.to(device)

# Train the network on GPU
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
for epoch in range(15):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        start_time = time()
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # Remember that you will have to send the inputs and targets at every step to the GPU too
        inputs, labels = inputs.to(device), labels.to(device)

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 200 == 199:    # print every 200 mini-batches
            print('[{}, {}] loss: {:.3f} time: {:.3f}:'.format
                  (epoch + 1, i + 1, running_loss / 200, (time()-start_time)*200))
            running_loss = 0.0

print('Finished Training')
```

```python
# Testing on GPU

correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data
        images, labels = images.to(device), labels.to(device)

        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

print('Accuracy of the network on the 10000 test images: %d %%' % (
    100 * correct / total))
```

## CAM

```python
from torchvision.io.image import read_image
from torchvision.transforms.functional import normalize, resize, to_pil_image
from torchvision.models import resnet18
#from torchcam.cams import CAM
from torchcam.methods import SmoothGradCAMpp
import matplotlib.pyplot as plt
from torchcam.utils import overlay_mask
```

```python
# Define your model
model = resnet18(pretrained=True).eval()
```

```python
# Set your CAM extractor
#cam_extractor = CAM(model, 'layer4', 'fc')
cam_extractor = SmoothGradCAMpp(model, 'layer4')
```

```python
# Get your input
img = read_image("../data/CNN_DNN/fox.jpg")
```

```python
# Preprocess it for your chosen model
input_tensor = normalize(
    resize(img, (224, 224)) / 255.,
    [0.485, 0.456, 0.406],
    [0.229, 0.224, 0.225]
    )
```

```python
# Preprocess your data and feed it to the model
out = model(input_tensor.unsqueeze(0))
# Retrieve the CAM by passing the class index and the model output
activation_map = cam_extractor(out.squeeze(0).argmax().item(), out)
```

```python
# Resize the CAM and overlay it
result = overlay_mask(
    to_pil_image(img),
    to_pil_image(activation_map[0], mode='F'),
    alpha=0.5
    )
```

```python
# Display it
plt.imshow(result)
plt.axis('off')
plt.tight_layout()
plt.show()
```
